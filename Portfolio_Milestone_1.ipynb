# Imports 

We've imported cv2 which is python variant of OpenCV. It comes preinstalled on the colab instance. You could download to local machine via "!pip install opencv-python" in a code cell or simply "pip install opencv-python" from command line, either directly to the machin or to virtual environment. 

```python
import cv2
from google.colab.patches import cv2_imshow
import requests
%matplotlib inline
```

# Setting Constants

1. I copied the link to the image from the critical thinking assignment
2. I set a name for the output file
3. I couldn't find an easily managable way to import the file to an opencv readable format without first writing the image to disk. So I have an intermediate written image via IO python commands 



```python
input_image_url = "https://frostlor-cdn-prod.courses.csuglobal.edu/lor/resources/src/7985e0c1-89b3-35e6-b709-8c90e828356c/shutterstock93075775--250.jpg"
output_filename = "AI Brain Image.jpg"
intermediate_filename = 'intermediate.jpg'
```

# Download file to intermediate file in disk

```python
response = requests.get(input_image_url, allow_redirects=True)
open(intermediate_filename, 'wb').write(response.content)
```

# Read file to OpenCV image object

```python
image =  cv2.imread(intermediate_filename)
```

# Show Image via OpenCV function
- Colab will not recognize opencv.imshow but the documentation from colab (error message) says to use cv2_imshow from colab patch

```python
cv2_imshow(image)
```

# Write image to disk 
- This writes the image to a file named: AI Brain Image.jpg 

```python
cv2.imwrite(output_filename,image)
```
