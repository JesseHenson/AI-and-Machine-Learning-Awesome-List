```python
from skimage import io 
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
import cv2
```

```python
image = io.imread("https://frostlor-cdn-prod.courses.csuglobal.edu/frost/assets/images/d46f8602-a3a9-38a7-be73-46c5cdf95933/Mod4CT1.jpg")
bgr_image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
cv2_imshow(bgr_image)
```

- Apply mean, median, and Gaussian filters using a 3x3 kernel. Additionally, for Gaussian, select two different values of sigma. Think about how to select good values of sigma for optimal results.
- Apply mean, median, and Gaussian filters using a 5x5 kernel. For Gaussian, use the same values of sigma you selected in the above step.
- Apply mean, median, and Gaussian filters using a 7x7 kernel. For Gaussian, use the same values of sigma you selected in the above step.
- Output your filter results as 3 x 4 side-by-side subplots to make comparisons easy to inspect visually. That is, your subplot should have 3 rows (1 for each kernel size) and 4 columns (1 for each filter type, 2 for Gaussian). Be sure to include row and column labels.

```python
sigma_1 = 3
sigma_2 = 20

images_obj = {
"mean_3": cv2.blur(image, (3,3)),
"median_3":cv2.medianBlur(image, 3),
"gaussian_3_1":cv2.GaussianBlur(image, (3,3),sigma_1),
"gaussian_3_2":cv2.GaussianBlur(image, (3,3),sigma_2),
"mean_5": cv2.blur(image, (5,5)),
"median_5":cv2.medianBlur(image,5),
"gaussian_5_1":cv2.GaussianBlur(image, (5,5),sigma_1),
"gaussian_5_2":cv2.GaussianBlur(image, (5,5),sigma_2),
"mean_7": cv2.blur(image, (7,7)),
"median_7":cv2.medianBlur(image, 7),
"gaussian_7_1":cv2.GaussianBlur(image, (7,7),sigma_1),
"gaussian_7_2":cv2.GaussianBlur(image, (7,7),sigma_2)
}

fig = plt.figure(figsize=(25,15))
count = 1

for key, value in images_obj.items():
  ax = fig.add_subplot(3,4, count)
  ax.xaxis.set_visible(False)
  ax.yaxis.set_visible(False)
  plt.imshow(value)
  plt.title(key)
  count += 1
```
