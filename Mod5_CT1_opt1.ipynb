```python
from nltk.corpus import wordnet
import nltk
nltk.download('wordnet')

dog = wordnet.synset('dog.n.01')
print(dog.lemma_names())

```

```python
def get_synonym(element):
  wordnet_words = wordnet.synsets(element)
  # print(wordnet_words)
  if wordnet_words != []:
    return wordnet_words[0].lemmas()[0].name()
  else: 
    return 

df['synomnym'] = df.vocab.apply(get_synonym)
```

```python
import random

def remove_random_word(element):
  word_arr = element.split()
  # print(len(word_arr))
  if len(word_arr) > 5:
    remove_arr = np.random.choice(word_arr, 3, replace=False) 
    for word in remove_arr:
      word_arr.remove(word)
  # print(len(word_arr))
  return " ".join(word_arr)

df['def_augment'] = df['def'].apply(remove_random_word)
df.head()
```
