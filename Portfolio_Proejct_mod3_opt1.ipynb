```python
from skimage import io
import matplotlib.pyplot as plt
import numpy as np
from google.colab.patches import cv2_imshow
import cv2
```

```python
my_hs = io.imread('https://scontent.fapa1-1.fna.fbcdn.net/v/t31.18172-8/14107620_10154525216530087_3742098245114040286_o.jpg?_nc_cat=110&ccb=1-5&_nc_sid=09cbfe&_nc_ohc=Ks-xS3dNYmwAX_p7-CQ&_nc_ht=scontent.fapa1-1.fna&oh=93c02586f59cb7111c9f55136c570c7a&oe=617F1E7F')
```

```python
my_hs_bgr = cv2.cvtColor(my_hs, cv2.COLOR_RGB2BGR)
my_hs_resize = cv2.resize(my_hs_bgr, (400,400))
h,w,c = my_hs_resize.shape
center = (int(w * .47),int(h*.15))
axesLength = (25, 35)
angle = 0
startAngle = 0
endAngle = 360
color = (0, 255, 0)
thickness = 2
my_hs_wcircle = cv2.ellipse(my_hs_resize, center, axesLength,
           angle, startAngle, endAngle, color, thickness)
right_start_point = (int(center[0]-3), int(center[1]-5)) 
left_start_point = (int(center[0]+ 3), int(center[1]-5)) 
right_end_point = (int(center[0]-15),int(center[1]+5))
left_end_point = (int(center[0]+15),int(center[1]+5))
color = (0, 0, 255)
thickness = 1
my_hs_right_eye = cv2.rectangle(my_hs_wcircle, right_start_point, right_end_point, color, thickness)
my_hs_eye = cv2.rectangle(my_hs_right_eye, left_start_point, left_end_point, color, thickness)

font = cv2.FONT_HERSHEY_SIMPLEX
org = (50, 25)
fontScale = 1
color = (255, 0, 0)
thickness = 2
my_hs_final = cv2.putText(my_hs_eye, 'This is me', org, font, 
                   fontScale, color, thickness, cv2.LINE_AA)
   

cv2_imshow(my_hs_final)
```
