```python
import numpy as np
import cv2 
from google.colab.patches import cv2_imshow


image = cv2.imread("Fig1.png")
```

```python
cv2_imshow(image)
```

```python
grey_image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
clahe = cv2.createCLAHE(clipLimit=4.0, tileGridSize=(3,3))
cl1 = clahe.apply(grey_image)
cv2_imshow(cl1)
```

```python
upper_a = image[110:245,110:190]
cv2_imshow(upper_a)
```

```python
upper_b = image[110:245,470:565]
cv2_imshow(upper_b)
```

```python
upper_c = image[110:210,870:945]
cv2_imshow(upper_c)
```

```python
lower_a = image[515:625,110:190]
cv2_imshow(lower_a)
```

```python
lower_b = image[480:610,470:565]
cv2_imshow(lower_b)
```

```python
lower_c = image[495:610,870:945]
cv2_imshow(lower_c)
```

```python
images = [lower_a, lower_b, lower_c, upper_a, upper_b, upper_c]

for img in images:
  resize_img = cv2.resize(img, (70,150),cv2.INTER_CUBIC)
  grey_img = cv2.cvtColor(resize_img, cv2.COLOR_BGR2GRAY)
  normalized_img = cv2.equalizeHist(grey_img)
  cv2_imshow(normalized_img)
```

```python
for img in images:
  resize_img = cv2.resize(img, (70,150),cv2.INTER_CUBIC)
  grey_img = cv2.cvtColor(resize_img, cv2.COLOR_BGR2GRAY)
  clahe = cv2.createCLAHE(clipLimit=4.0, tileGridSize=(3,3))
  cl1 = clahe.apply(grey_img)
  cv2_imshow(cl1)
```

```python
import matplotlib.pyplot as plt


figure, axes = plt.subplots(nrows=2, ncols=6)
figure.set_size_inches(14,8)

for index, img in enumerate(images):


  resize_img = cv2.resize(img, (70,150),cv2.INTER_CUBIC)
  grey_img = cv2.cvtColor(resize_img, cv2.COLOR_BGR2GRAY)
  axes[0,index].hist(grey_img)
  clahe = cv2.createCLAHE(clipLimit=4.0, tileGridSize=(3,3))
  cl1 = clahe.apply(grey_img)
  axes[1,index].hist(cl1)
  # cv2_imshow(cl1)
figure.tight_layout()
```
