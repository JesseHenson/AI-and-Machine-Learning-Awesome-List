# Goal

- Split dog picture into color channels 
- Switch red channel to green channel and vice versa

```python
import cv2
import numpy as np
import requests
from google.colab.patches import cv2_imshow

%matplotlib inline



# intermediate_file = "intermediate_file"
# output_file = "GRB Dog"
# input_image_url = "https://frostlor-cdn-prod.courses.csuglobal.edu/lor/resources/src/89f79919-379b-3a8f-997a-98f3dd1d3a8a/shutterstock215592034--250.jpg"

import matplotlib.pyplot  as plt
from skimage import io

image=io.imread ('https://frostlor-cdn-prod.courses.csuglobal.edu/lor/resources/src/89f79919-379b-3a8f-997a-98f3dd1d3a8a/shutterstock215592034--250.jpg')
img_bgr = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
plt.imshow(image)
cv2.imshow("image BRG", img_bgr)
```

# Show images in one plot using subplot

- imshow shows vertical images, this shows side by side using subplot api from matplotlib

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(nrows = 1, ncols = 3, figsize = (15,5))
 
for c, ax in zip(range(3), ax):
  split_img = np.zeros(image.shape, dtype = 'uint8')
  split_img[:, :, c] = image[:, :, c]
  ax.imshow(split_img)
 
plt.show()
```

# use array swap syntax to swap red channel with Green

- Since opencv reads in images in BGR rather than RGB. Blue then green the red will be the three channels. 
- I will swap channel 1 and channel 2

```python
copy_image = img_bgr.copy()
r = copy_image[:,:,2]
g = copy_image[:,:,1] 
b = copy_image[:,:,0]

full_copy = np.zeros([257,250,3])

full_copy[:,:,0] = b
full_copy[:,:,1] = r
full_copy[:,:,2] = g
cv2_imshow(full_copy)
```

```python
full_copy_image = full_copy.astype(np.uint8)
```

# histogram of image color 



```python
color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv2.calcHist([img_bgr],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.show()
```

```python
color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv2.calcHist([full_copy_image],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.show()
```
